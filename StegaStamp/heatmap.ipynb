```python
import numpy as np
import matplotlib.pyplot as plt

test_type = 'solo'
img_name = 'test2'
ext = '.png'
size = (5,5)

try:
    # red_heatmap = np.load(f'heatmaps/red_heatmap_solo_{img_name}.npy')
    # green_heatmap = np.load(f'heatmaps/green_heatmap_solo_{img_name}.npy')
    # blue_heatmap = np.load(f'heatmaps/blue_heatmap_solo_{img_name}.npy')
    # total_heatmap = np.load(f'heatmaps/total_heatmap_solo_{img_name}.npy')
    red_heatmap = np.load(f'heatmaps/red_heatmap_{test_type}_{img_name}.npy')
    green_heatmap = np.load(f'heatmaps/green_heatmap_{test_type}_{img_name}.npy')
    blue_heatmap = np.load(f'heatmaps/blue_heatmap_{test_type}_{img_name}.npy')
    total_heatmap = np.load(f'heatmaps/total_heatmap_{test_type}_{img_name}.npy')
    print('Successfully loaded!')
    print(red_heatmap.shape)
except Exception as e:
    print(e)
    print('Failed to load!')
```

```python
red_heatmap = np.var(red_heatmap, axis=0)
green_heatmap = np.var(green_heatmap, axis=0)
blue_heatmap = np.var(blue_heatmap, axis=0)
total_heatmap = np.var(total_heatmap, axis=0)
```
